```python
#Sending a message to multiple clients using for loops
for number in range(3):
    print("Hi, this is your first message", number + 1)
```

```python
#Using Nested loop to print points in a cartesian plain
for x in range(5):
    for y in range(3):
        print((x, y))
```

```python
#Iterables
print(type(range(5)))
```

```python
for x in range(3):
    print(x)
```

```python
for x in "Isaac":
    print(x)
```

```python
for x in ["Mango", "Orange", "Kiwi"]:
    print(x)
```

```python
#While loops to repeat a process as long as the condition is true
number = 500
while number > 0:
    print(number)
    number //= 2
```

```python
numbers = [1, 2, 3, 4, 5]
condition = 0
while condition < len(numbers):
    print(f"{condition} is less than {len(numbers)}")
    condition += 1
```

```python
#Infinite loops
number = 5
while number > 0:
    print("True")
```

```python
#Exercise: Write program to display the even numbers between 1-10
number_list = []
for number in range(1, 10):
    if number % 2 == 0:
        print(number)
        number_list.append(number)
print(f"We have {len(number_list)} even numbers")
    
```
